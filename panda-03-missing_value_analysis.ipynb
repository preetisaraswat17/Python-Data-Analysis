```python
#Nan value could be anything so they are neither false, nor true and neither equal to each other 
#first make a dataframe
df=pd.DataFrame({
        'a':[1,2,3,4],
        'val':['apples','oranges','grapes','mango']})
        
```

```python
df
```

```python
#make a second dataframe
new_data=pd.DataFrame({
        'a1':[0,3,5],
        'val1':['foo','bar','fig']})
```

```python
#merge both data
my_data=df.merge(new_data,left_on='a', right_on='a1',how='outer')
```

```python
my_data
```

```python
#check for null values in the file by isnull command
pd.isnull(my_data)
```
