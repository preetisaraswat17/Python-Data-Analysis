```python
data2=pd.read_csv(r'C:\Users\PREETI SARASWAT\Desktop\test\scipy-2017-tutorial-pandas-master\scipy-pandas\data\weather.csv',\
               delimiter=',')
```

whenever you see dataset with 2 or more same rows just a few changes, thats a problem as its has lots of duplicate data and we need to remove this issue.

```python
data2.head()
```

```python
#like in this data..we have same rows except for each tmax and tmin..so
#step 1) melt the column
weather=pd.melt(data2, id_vars=['id','year','month','element'], var_name='day', value_name='temp')
```

```python
weather.head()
```

```python
#step2) pivot up the data:to pivot is to turn the data to view it from different perspectives., 
#so what we need to do is to basically delete the element column and instead add 2 columns with tmax and tmin 
```

```python
weather_clean=weather.pivot_table( index=['id','year','month','day'], columns='element', values='temp')
#index= columns that we need to be same
#columns=values that we want to pivot
#values= that we want to fill in the pivot columns
```

```python
weather_clean.head(20)
```

```python
weather_clean=weather_clean.reset_index().head(10)
weather_clean
```

```python
#Step3) data is in much better condition now, still we have these values in day column that doesnt make much sense. 
#so we can just slice it upto into integer values and remove the d part, for this you can use lambda function

day_var=weather_clean['day'].apply(lambda x: x[1:])
```

```python
#fill the day column with new values
weather_clean['day']=day_var
```
